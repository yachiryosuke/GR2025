# 6. 2次元ベクトルのスケーリング

## 6.1 ライブラリのインポート

```python
import numpy as np
import math
import os
from tqdm import tqdm
```

## 6.2 関数map_rangeを定義

```python
def map_range(x, a, b, c, d):
    """
    変数 $x$ を 範囲 [$a$, $b$] から [$c$, $d$] にリマップします
    :param x: 変換する対象
    :param a: 変換前の下端
    :param b: 変換前の上端
    :param c: 変換後の下端
    :param d: 変換後の上端
    :return: リマップした値
    """
    return (x - a) / (b - a) * (d - c) + c
```

## 6.3 2次元ベクトルのスケーリング

```python
model_name = "prott5"  # モデル名
for fold in range(1, 6):
    method_name = "pcaotsne2_cog"  # 使った次元削減の手法
    dir_name = "fold_pcaotsne2_cog_ppx5_cosine" # データが格納されている親ディレクトリ

    train_vecs = f"./data/embedding-vectors_exclusion/{model_name}/{dir_name}/fold{fold}_train_{method_name}" # trainデータ
    train_files = [os.path.join(train_vecs, f) for f in os.listdir(train_vecs) if f.endswith(".npy")]  # .npy ファイルを全て取得
    train_output_dir = f"./data/embedding-vectors_exclusion/{model_name}/{dir_name}/fold{fold}_train_{method_name}_scaled_10_170_ut"
    os.makedirs(train_output_dir, exist_ok=True)

    test_vecs = f"./data/embedding-vectors_exclusion/{model_name}/{dir_name}/fold{fold}_test_{method_name}" # testデータ
    test_files = [os.path.join(test_vecs, f) for f in os.listdir(test_vecs) if f.endswith(".npy")]  # .npy ファイルを全て取得
    test_output_dir = f"./data/embedding-vectors_exclusion/{model_name}/{dir_name}/fold{fold}_test_{method_name}_scaled_10_170_ut"
    os.makedirs(test_output_dir, exist_ok=True)
   
    for filename in tqdm(train_files):
        old_vec = np.load(filename, mmap_mode="r")
        new_vec = np.zeros(old_vec.shape)
        h = old_vec.shape[0]

        for i in range(h):
            x = old_vec[i][0]
            y = old_vec[i][1]
            #z = x + 1j * y
            r = math.sqrt(x**2 + y**2)
            rad = math.atan2(y,x)
            deg = math.degrees(rad)
            if deg < -90:
                deg += 360
            new_deg = 4 / 9 * deg + 50
            new_theta = math.radians(new_deg)
            #new_theta = np.radians(map_range(theta, -90, 270, 10, 170))
            new_x = r * math.cos(new_theta)
            new_y = r * math.sin(new_theta)
            new_vec[i] = [new_x, new_y]

        save_filename = os.path.basename(filename).replace(method_name, f"{method_name}_scaled_10_170")
        np.save(os.path.join(train_output_dir, save_filename), new_vec)

    for filename in tqdm(test_files):
        old_vec = np.load(filename, mmap_mode="r")
        new_vec = np.zeros(old_vec.shape)
        h = old_vec.shape[0]

        for i in range(h):
            x = old_vec[i][0]
            y = old_vec[i][1]
            #z = x + 1j * y
            r = math.sqrt(x**2 + y**2)
            rad = math.atan2(y,x)
            deg = math.degrees(rad)
            if deg < -90:
                deg += 360
            new_deg = 4 / 9 * deg + 50
            new_theta = math.radians(new_deg)
            #new_theta = np.radians(map_range(theta, -90, 270, 10, 170))
            new_x = r * math.cos(new_theta)
            new_y = r * math.sin(new_theta)
            new_vec[i] = [new_x, new_y]

        save_filename = os.path.basename(filename).replace(method_name, f"{method_name}_scaled_10_170")
        np.save(os.path.join(test_output_dir, save_filename), new_vec)
   
```
