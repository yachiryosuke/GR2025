# 5. t-SNEを用いた2段階目の次元削減（N → 2）

## 5.1 ライブラリのインポート

```python
import numpy as np
from openTSNE import TSNE
import os
import re
from tqdm import tqdm
```

## 5.2 t-SNEの実行

```python
model_name = "protbert"
perplexity = 5
metric = "cosine"

for fold in range(1, 6):
    # trainデータの読み込み
    train_dir = f"./data/embedding-vectors_exclusion/{model_name}/fold_pca_cog/fold{fold}_train_pca_cog"  # PCAデータの読み込み
    train_files = [os.path.join(train_dir, f) for f in os.listdir(train_dir) if f.endswith(".npy")]  # .npy ファイルを全て取得
    train_files = sorted(train_files, key=lambda p: int(re.search(r'(\d+)_pca.npy$', p).group(1)))
    print(train_files)

    # trainデータのリストの作成
    train_vecs_lst = []
    train_len_lst = []
    for filename in tqdm(train_files):
        arr = np.load(filename, mmap_mode="r")
        train_vecs_lst.append(arr)
        train_len_lst.append(len(arr))

    # testデータの読み込み
    test_dir = f"./data/embedding-vectors_exclusion/{model_name}/fold_pca_cog/fold{fold}_test_pca_cog"  # PCAデータの読み込み
    test_files = [os.path.join(test_dir, f) for f in os.listdir(test_dir) if f.endswith(".npy")]  # .npy ファイルを全て取得
    test_files = sorted(test_files, key=lambda p: int(re.search(r'(\d+)_pca.npy$', p).group(1)))
    print(test_files)

    # testデータのリストの作成
    test_vecs_lst = []
    test_len_lst = []
    for filename in tqdm(test_files):
        arr = np.load(filename, mmap_mode="r")
        test_vecs_lst.append(arr)
        test_len_lst.append(len(arr))

    # OpenTSNEのfit部
    train_vecs = np.vstack(train_vecs_lst)
    test_vecs = np.vstack(test_vecs_lst)
    opentsne = TSNE(n_components=2, metric=metric, perplexity=perplexity, random_state=42)
    reducer = opentsne.fit(train_vecs)

    # OpenTSNEのtransform済みtrainデータの保存
    train_output_dir = f"./data/embedding-vectors_exclusion/{model_name}/fold_pcaotsne2_cog_ppx{perplexity}_{metric}/fold{fold}_train_pcaotsne2_cog"
    os.makedirs(train_output_dir, exist_ok=True)
    print(f"fold{fold} train_vecs transform")
    train_vecs_trans = reducer.transform(train_vecs)

    start = 0
    idx = 0

    for filename in tqdm(train_files):
        end = start + train_len_lst[idx]
        coord = train_vecs_trans[start:end]
        save_filename = os.path.basename(filename).replace("pca", "pcaotsne2")
        np.save(os.path.join(train_output_dir, save_filename), np.array(coord))
        start = end
        idx += 1

    # UMAPのtransform済みtestデータの保存
    test_output_dir = f"./data/embedding-vectors_exclusion/{model_name}/fold_pcaotsne2_cog_ppx{perplexity}_{metric}/fold{fold}_test_pcaotsne2_cog"
    os.makedirs(test_output_dir, exist_ok=True)
    print(f"fold{fold} test_vecs transform")
    test_vecs_trans = reducer.transform(test_vecs)

    start = 0
    idx = 0

    for filename in tqdm(test_files):
        end = start + test_len_lst[idx]
        coord = test_vecs_trans[start:end]
        save_filename = os.path.basename(filename).replace("pca", "pcaotsne2")
        np.save(os.path.join(test_output_dir, save_filename), np.array(coord))
        start = end
        idx += 1
    
```
