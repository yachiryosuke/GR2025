# 3. 重心の位置ベクトル群の算出

## 3.1 ライブラリのインポート

```python
import os
import numpy as np
from tqdm import tqdm
import re
```

## 3.2 データの読み込み

```python
model_name = "prott5"

for fold in range(1, 6):
    data_dir = f"./data/embedding-vectors_exclusion/{model_name}/base/fold{fold}_train"  # foldごとの1024次元のデータの読み込み
    files = [os.path.join(data_dir, f) for f in os.listdir(data_dir) if f.endswith(".npy")]  # .npy ファイルを全て取得
    files = sorted(files, key=lambda p: int(re.search(r'(\d+).npy$', p).group(1)))

    lst_cog = [] #ベクトルの重心のリスト
    for filename in tqdm(files):
        arr = np.load(filename, mmap_mode="r")  # mmap_mode により 一気に読み込むことを防止できる
        cumsum = arr.T.tolist()
        cog = [] #個々のファイルの重心ベクトルを格納
        for i in range(len(cumsum)):
            for j in range(1, len(cumsum[i])):
                cumsum[i][j] += cumsum[i][j - 1]
            cog.append(sum(cumsum[i]) / len(cumsum[i]))
        lst_cog.append(cog)

    arr_cog = np.array(lst_cog)

    output_dir = f"./data/embedding-vectors_exclusion/{model_name}/cog1024"
    os.makedirs(output_dir, exist_ok=True)

    np.save(os.path.join(output_dir,f"fold{fold}_cog.npy"), arr_cog)
    
```
